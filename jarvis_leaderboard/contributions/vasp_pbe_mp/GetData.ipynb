```python
import glob,os
for i in glob.glob('../vasp_optb88vdw/SinglePropertyPrediction-test-bandgap_JVASP_*-dft_3d-ES-mae.csv.zip'):
    jid='JVASP-'+i.split('_')[3]
    jid=int(i.split('_')[3])
    mp=df[df['JV#']==jid]['MP'].values[0]
    opt=df[df['JV#']==jid]['OPT'].values[0]
    hse=df[df['JV#']==jid]['HSE06'].values[0]
    aflow=df[df['JV#']==jid]['AFLOW'].values[0]
    oqmd=df[df['JV#']==jid]['OQMD'].values[0]
    if mp!='-':
        fname=i.split('../vasp_optb88vdw/')[1].split('.zip')[0]
        f=open(fname,'w')
        line='id,prediction\n'
        f.write(line)
        line='JVASP-'+str(jid)+','+str(mp)+'\n'
        f.write(line)
        f.close()
        cmd = 'zip '+fname+'.zip '+fname
        os.system(cmd)
        cmd = 'rm '+fname
        os.system(cmd)
    #print (jid)
    print (jid,mp,opt,hse,aflow,oqmd)
    #break
```
